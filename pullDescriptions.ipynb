```python
import os
import openai
import gzip
import json
import pickle
from pathlib import Path


def save_json(obj: dict, file):
    file = _file_ending(file, "json")
    create_directories(file)
    with open(file, 'a') as f:
        json.dump(obj, f, indent=4)

def create_directories(path):
    Path(os.path.dirname(path)).mkdir(parents=True, exist_ok=True)
          
def _file_ending(file, ending):
    return f"{file}.{ending}" if f".{ending}" not in file else file

def getDescriptions(methods):
    openai.api_key = ""
    n=20
    #1380
    for i in range(1380, len(methods), n):
        print(f"{i}/{len(methods)}")
        try:
            code = [method['code']+"\n/* Explain what the previous function is doing: It" for method in methods[i:i+n]]
            response = openai.Completion.create(
            engine="code-davinci-001",
            prompt=code,
            temperature=0,
            max_tokens=128,
            top_p=1,
            frequency_penalty=0.6,
            presence_penalty=0
            )
            for j,method in  enumerate(methods[i:i+n]):
                method['description']="it" +response['choices'][j].text
            save_json(methods[i:i+n], r'C:\Users\vikram14\Desktop\Research\IntelliDiff\comp410\methods_w_descriptions')
        except Exception as e:
            print(e)
            print(f"error at {i}")
            continue
        #method['description']="it" + response.text
```
